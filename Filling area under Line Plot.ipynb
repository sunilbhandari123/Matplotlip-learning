```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('data/fill.csv.txt')
```

```python
ages=data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']
js_salaries=data['JavaScript']
```

```python
plt.plot(ages,py_salaries,color='#A020F0',label='Python')
plt.plot(ages,js_salaries,color='#FFC0CB',label='Javascript')
plt.legend()
plt.title('Salary of Developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.show()
```

```python
#To fill the area above the python we can use plt.fill_between()
```

```python
plt.plot(ages,py_salaries,color='#A020F0',label='Python')
plt.plot(ages,js_salaries,color='#FFC0CB',label='Javascript')
plt.fill_between(ages,py_salaries)
plt.legend()
plt.title('Salary of Developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.show()
```

```python
#Here the filling of the color is very low therefore to reduce we use alpha
```

```python
plt.plot(ages,py_salaries,color='#A020F0',label='Python')
plt.plot(ages,js_salaries,color='#FFC0CB',label='Javascript')
plt.fill_between(ages,py_salaries,alpha=0.25)
plt.legend()
plt.title('Salary of Developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.show()
```

```python
overall_median=57287
```

```python
#to fill above the median salary we use 
```

```python
plt.plot(ages,py_salaries,color='#A020F0',label='Python')
plt.plot(ages,js_salaries,color='#FFC0CB',label='Javascript')
plt.fill_between(ages,py_salaries,overall_median,alpha=0.25)
plt.legend()
plt.title('Salary of Developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.show()
```

```python
#for changing the color above and below the median salary
```

```python
plt.plot(ages,py_salaries,color='#A020F0',label='Python')
plt.plot(ages,js_salaries,color='#FFC0CB',label='Javascript')
plt.fill_between(ages,py_salaries,overall_median, where=(py_salaries>overall_median),interpolate=True,alpha=0.25)
plt.fill_between(ages,py_salaries,overall_median, where=(py_salaries<=overall_median),interpolate=True,alpha=0.25)
plt.legend()
plt.title('Salary of Developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.show()
```

```python
# to fill inbetween the two curves python and javascript
```

```python
plt.plot(ages,py_salaries,color='#A020F0',label='Python')
plt.plot(ages,js_salaries,color='#FFC0CB',label='Javascript')
plt.fill_between(ages,py_salaries,js_salaries, where=(py_salaries>js_salaries),interpolate=True,alpha=0.25)
plt.fill_between(ages,py_salaries,js_salaries, where=(py_salaries<=js_salaries),interpolate=True,alpha=0.25)
plt.legend()
plt.title('Salary of Developers')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.show()
```
