```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
x=[5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
```

```python
y=[7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
color=[7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
size=[251,582,256,365,245,425,325,425,512,265,456,456,1213,542,245,325,253,452,325,254]
```

```python
plt.scatter(x,y,s=100,c='green')
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

#s is size
#c is color
```

```python
#ADDING EDGES
```

```python
plt.scatter(x,y,s=100,c='green',edgecolor='black',linewidth=1,alpha=0.75)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#adding color by the list not staticially
```

```python
plt.scatter(x,y,s=100,c=color,edgecolor='black',cmap='Greens',linewidth=1,alpha=0.75)
#cmap is used to put the desire color when we are accessing the colors from the list
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#ADDING COLOR BAR
```

```python
plt.scatter(x,y,s=100,c=color,edgecolor='black',cmap='Greens',linewidth=1,alpha=0.75)
plt.title('Scatter Plot')
cbar=plt.colorbar()
cbar.set_label('Satisfaciton')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#ADDING SIZES
```

```python
#passing the size from the list not statically
```

```python
plt.scatter(x,y,s=size,c=color,edgecolor='black',cmap='Greens',linewidth=1,alpha=0.75)
plt.title('Scatter Plot')
cbar=plt.colorbar()
cbar.set_label('Satisfaciton')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
