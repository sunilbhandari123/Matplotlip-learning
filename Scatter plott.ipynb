```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('data/scatter.csv.txt')view_count,likes,ratio
```

```python
view_count=data['view_count']
likes=data['likes']
ratio=data['ratio']
```

```python
plt.scatter(view_count,likes,edgecolor='black',linewidth=1)
plt.title('Trending Youtube Videos')
plt.xlabel('Total Count')
plt.ylabel('Total Likes')

plt.show()
```

```python
# HERE THE PLOT IS NOT SO GOOD WE CANNOT SE ALL THE PLOT TO SEE ALL THE PLOT WE USE THE LOG LIKE BELOW
```

```python
plt.scatter(view_count,likes,edgecolor='black',linewidth=1)
plt.title('Trending Youtube Videos')
plt.xlabel('Total Count')
plt.ylabel('Total Likes')
plt.xscale('log')
plt.yscale('log')

plt.show()
```

```python
#now the plots seems good and easy to view
```

```python
#HERE RATIO CAN BE USED AS THE SIZE OR COLOR BUT IN THIS CASE THE COLORS WILL BE GOOD SO I USE RATIO AS COLOR
```

```python
plt.scatter(view_count,likes,edgecolor='black',c=ratio,linewidth=1)
plt.title('Trending Youtube Videos')
plt.xlabel('Total Count')
plt.ylabel('Total Likes')
plt.xscale('log')
plt.yscale('log')

plt.show()
```

```python
#adding the color bar||
```

```python
plt.scatter(view_count,likes,edgecolor='black',c=ratio,linewidth=1)
cbar=plt.colorbar()
cbar.set_label('Views/Like Ratio')
plt.title('Trending Youtube Videos')
plt.xlabel('Total Count')
plt.ylabel('Total Likes')
plt.xscale('log')
plt.yscale('log')

plt.show()
```
