```python
from matplotlib import pyplot as plt
```

```python
numbers=[59234,54234,45234,32234,23345]
```

```python
labels=['Python','javascript','html/css','java','c++']
colors=['#FFC0CB','#ADD8E6','#90EE90','#CBC3E3','#C4A484']
explode=[0.08,0,0,0,0]
```

```python
plt.pie(numbers,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'})
#wedges properties is used to put the seperater inbetween

plt.title('Pie Chart')
plt.show()
```

```python
#pie chart doesnot look good for the more number of data they are good for upto five numbers of data
```

```python
#piechart is perfered when the data is less than or equal to five
```

```python
#EXPLODE
```

```python
plt.pie(numbers,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'},explode=explode)
plt.title('Pie Chart')
plt.show()
```

```python
#OTHER ARGUMENT THAT WE CAN PASS IN THE PIECHARTS
```

```python
#SHADOW
```

```python
plt.pie(numbers,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'},explode=explode,shadow=True)
plt.title('Pie Chart')
plt.show()
```

```python
#STARTING ANGLE
```

```python
plt.pie(numbers,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'},explode=explode,startangle=90)
plt.title('Pie Chart')
plt.show()
```

```python
#last but not least display percentage
```

```python
plt.pie(numbers,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'},explode=explode,autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
