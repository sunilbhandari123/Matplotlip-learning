```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('data/timeseries.csv.txt')
```

```python
data["Date"]=pd.to_datetime(data['Date'])   
data.sort_values('Date',inplace=True)



price_date=data['Date']
price_close=data['Close']
```

```python
plt.plot_date(price_date,price_close,linestyle='solid')
plt.title('Bitcoin Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
#here the date is not visible so use should change it to the inclind
```

```python
plt.plot_date(price_date,price_close,linestyle='solid')
plt.gcf().autofmt_xdate()
plt.title('Bitcoin Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```
