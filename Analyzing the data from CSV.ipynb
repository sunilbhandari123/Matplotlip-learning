```python
#for printing the bar chat in the csv file
language_counter=Counter()
```

```python
for response in lang_responses:
    language_counter.update(response.split(';'))
```

```python
print(language_counter)
```

```python
print(language_counter.most_common(15))  #it will show the language with the most users
```

```python
#For printing  in bargraph we should make the language in axis and users in another
#we can do it by following
```

```python
languages=[]
popularity=[]
```

```python
for item in (language_counter.most_common(15)):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
plt.bar(languages,popularity)
```

```python
#in the above graph if the number of data in x axis it is hard to read so we perfered the horizental bargraph
```

```python
languages.reverse()  #it helps us to bring the highest first such that it is easy to read
popularity.reverse()
plt.barh(languages,popularity)  #barh is used for horizental bargraph
plt.title('Most Popular Language')
plt.xlabel('Number of people using')
plt.tight_layout()  #spacing
```
