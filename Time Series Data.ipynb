```python
import pandas as pd
```

```python
from  matplotlib import pyplot as plt
from datetime import datetime,timedelta
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('seaborn')
```

```python
dates=[
    datetime(2023,2,6),
    datetime(2023,2,7),
    datetime(2023,2,8),
    datetime(2023,2,9),
    datetime(2023,2,10),
    datetime(2023,2,11),
    datetime(2023,2,12)
]
```

```python
y=[0,1,3,4,6,5,7]
```

```python
plt.plot_date(dates,y)
plt.show()
```

```python
#to draw the line on the points
```

```python
plt.plot_date(dates,y,linestyle='solid')
plt.show()
```

```python
#here the date is looking not good if it is inclined then it will look nice for that
```

```python
plt.plot_date(dates,y,linestyle='solid')
plt.gcf().autofmt_xdate()
plt.show()
```

```python
#Now it looks nice
```

```python
# TO FORMAT THE DATE WE USE
```

```python
plt.plot_date(dates,y,linestyle='solid')
plt.gcf().autofmt_xdate()
date_format=mpl_dates.DateFormatter('%b,%d %Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.show()
```
